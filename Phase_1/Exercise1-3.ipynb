```python
import openai
from IPython.display import display, Markdown
import getpass
import os

# 1. Load API Key
try:
    from dotenv import load_dotenv
    load_dotenv()
except ImportError:
    pass

api_key = os.getenv('OPENAI_API_KEY')
if not api_key:
    api_key = getpass.getpass('Enter your OpenAI API Key: ')
    os.environ['OPENAI_API_KEY'] = api_key

client = openai.OpenAI(api_key=api_key)
print("OpenAI Client configured.")
```

```python
def analyze_contracts(contract_a, task_description, contract_b=None, specific_instructions=""):
    """
    Uses AI to compare or analyze contracts.
    Includes retry logic to handle potential connection errors.
    """
    if not contract_a or len(contract_a.strip()) < 10:
        print("Error: Please provide detailed text for Contract A.")
        return

    print(f"Processing: {task_description}...")
    
    # Construct Prompt
    documents_section = f"""
    >>> CONTRACT A:
    {contract_a}
    """
    
    if contract_b and len(contract_b.strip()) > 10:
        documents_section += f"""
        
        >>> CONTRACT B:
        {contract_b}
        """
    
    prompt = f"""
    System: You are an expert legal analyst specializing in contract law and risk assessment.
    
    Source Documents:
    {documents_section}
    
    Task Goal: {task_description}
    
    Quality Control Requirements:
    - Maintain strict client confidentiality.
    - Identify both subtle and obvious legal risks.
    - Provide objective, fact-based comparisons.
    
    Specific Instructions: {specific_instructions}
    
    Return the result in Markdown format with professional headers.
    """
    
    max_retries = 3
    import time
    
    for attempt in range(max_retries):
        try:
            response = client.chat.completions.create(
                model='gpt-4o',
                messages=[{'role': 'user', 'content': prompt}]
            )
            result = response.choices[0].message.content
            display(Markdown(result))
            return # Success
        except Exception as e:
            if attempt < max_retries - 1:
                print(f"Attempt {attempt + 1} failed: {e}. Retrying in 2 seconds...")
                time.sleep(2)
            else:
                print(f"Error: Failed after {max_retries} attempts. Final error: {e}")
```
