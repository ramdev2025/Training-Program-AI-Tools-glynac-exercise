LEGAL ANALYSIS

1. Federal Arbitration Act (FAA) — general rule
   The FAA establishes a strong federal policy favoring enforcement of arbitration agreements and directs courts to enforce valid arbitration clauses unless a statutory or common-law ground for revocation applies. ([Legal Information Institute][1])

2. Separability / who decides validity of the contract vs. the clause
   Under the separability doctrine, challenges to the contract’s validity generally go to the arbitrator unless the challenge is directed specifically at the arbitration clause itself. This principle frequently requires courts to send disputes to arbitration while leaving certain threshold questions to the arbitrator. ([Justia Law][2])

3. Delegation clauses — who decides arbitrability
   If the contract contains an express delegation provision (“arbitrator decides arbitrability”) or incorporates arbitral rules that empower the arbitrator to rule on arbitrability, courts will generally honor that delegation and decline to decide threshold arbitrability questions. Recent Supreme Court precedent enforces such delegations. This limits judicial tools to keep the dispute in court when delegations are present. ([Legal Information Institute][3])

4. Defenses and limits to enforceability
   Contract defenses remain available (fraud, duress, unconscionability, statutory preemption), but the Supreme Court has narrowed some state-law defenses—e.g., courts must apply the FAA’s standards and have upheld arbitration clauses against certain state-law attempts to invalidate them (e.g., class-action waiver contexts). Any challenge must be framed to attack the clause (or the delegation provision) specifically to remain in court. ([Justia Law][4])

5. Practical effect on remedies and process (arbitration characteristics)
   Arbitration tends to be faster and more private than ordinary litigation, with streamlined procedures and limited appellate review. It typically offers narrower discovery, finality of awards, and fewer procedural safeguards than court litigation—advantages and disadvantages that affect a small business seeking prompt relief for ongoing losses and property damage. ([cpradr.org][5])

COMPARISON: ARBITRATION VS LITIGATION (SHORT)

* Arbitration — Pros: potentially faster resolution; confidential proceedings; selection of subject-matter–experienced arbitrator. Cons: limited discovery, limited appellate review, arbitral fees (arbitrator/admin), possible difficulty securing emergency provisional relief quickly depending on rules and arbitrator availability. ([cpradr.org][5])

* Litigation — Pros: full discovery, injunctive relief and provisional remedies readily available from the court, appellate avenues. Cons: longer timeline, higher litigation costs, public record. (Given $2,000/day losses, delay imposes real financial pressure.)

1. Immediate work plan (non-waivable steps)

   b. Preserve and produce evidence: maintain contemporaneous project logs, photos, invoices, payment records, communications with Apex, and inspection estimates for floor repairs.
   c. Draft a response to the Motion to Dismiss that (i) denies enforceability if a strong basis exists, or (ii) requests conditions/expedited treatment if arbitration is compelled.

2. Litigation posture vs. arbitration posture — recommended approach

   * If clause is ambiguous, lacks a clear delegation provision, or has a plausible, specific challengeable defect (e.g., unconscionability, fabrication/forgery, or a carve-out for injunctive relief), resist arbitration and ask the court to retain jurisdiction—file a substantive opposition arguing arbitrability is for the court. This aligns with client goals: urgent relief, quicker access to provisional remedies, and superior discovery to quantify damages (lost revenue and floor replacement). (Action: oppose.) In that event, accept arbitration but demand protective measures: (i) seek an expedited schedule, (ii) request an arbitrator with construction/contract experience, (iii) preserve rights to seek emergency provisional relief from the court under FAA §§ 3–4 or otherwise as needed, and (iv) require bifurcation or advance briefing on interim remedies (or negotiate a tolling/standstill agreement over deadlines and remedies while arbitration proceeds). (Action: compel/arbitrate under protective terms.)

3. Tactical priorities given client profile and damages

The cost of delay at ~$2,000/day counsels in favor of the fastest effective forum for obtaining relief. If court access meaningfully increases speed to injunctive or provisional relief (e.g., to force remediation or secure funds for repairs), prioritize staying in court. If neither the clause nor the circumstances permit that, force an expedited arbitration with firm interim remedies. ([cpradr.org][5])

If the clause is enforceable and delegates arbitrability, pivot immediately to a tightly-managed, expedited arbitration strategy that preserves access to provisional court relief as needed.
